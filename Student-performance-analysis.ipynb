## Task-1: Student Performance Analysis
- Import necessary libraries : Pandas, numpy, metploatlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Load the Dataset

- The Student Performance dataset is loaded using pandas.
- The dataset uses semicolon (;) as a delimiter.

```python
df = pd.read_csv(r"D:\BE\Be Sem 8\Virtual Data Analyst\Task 1\student\student-mat.csv", sep=';')

# Display the first five rows using .head().
df.head()
```

## Step 2: Data Exploration

- In this step, we explore the structure of the dataset, data types,
and check for missing values.

```python
# Display the dataset's size using .shape.
df.shape
```

```python
# Display the data types of each column using .dtypes.
df.dtypes
```

```python
# Check for missing values using .isnull().sum().
df.isnull().sum()
```

## Step 3: Data Cleaning

- Missing numerical values are handled using median values using .median().
- Duplicate records are removed to improve data quality using .drop_duplicates().

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df = df.drop_duplicates()
```

## Step 4: Data Analysis

- In this step, we answer analytical questions based on the dataset.

```python
# Average final math score (G3)
average_g3 = df['G3'].mean()
average_g3
```

```python
# Students scoring above 15 in G3
students_above_15 = df[df['G3'] > 15].shape[0]
students_above_15
```

```python
# Correlation between study time and final grade
correlation = df['studytime'].corr(df['G3'])
correlation
```

```python
# Gender-wise average final grade
gender_avg = df.groupby('sex')['G3'].mean()
gender_avg
```

## Step 5: Data Visualization

- Graphs are used to visualize student performance and relationships
between different variables.

```python
# Histogram: distribution of final grades
plt.hist(df['G3'], bins=10)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

```python
# Scatter plot: study time vs final grade
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
# Bar chart: average G3 by gender
gender_avg.plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.title("Average Final Grade by Gender")
plt.show()
```
